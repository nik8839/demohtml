```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('insurance_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['age','affordibility']],df.bought_insurance,test_size=0.2,random_state=25)

```

```python
x_train.shape
```

```python
x_train_scale=x_train.copy()
x_train_scale['age']=x_train_scale['age']/100;
x_test_scale=x_test.copy()
x_test_scale['age']=x_test_scale['age']/100;
```

```python
from keras.engine.training import optimizer
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']


)
```

```python
model.fit(x_train_scale,y_train,epochs=5000)
```

```python
model.evaluate(x_test_scale,y_test)
```

```python
coef,intercept=model.get_weights()
coef,intercept
```

```python
#log loss
def log_loss(y_true,y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i,epsilon) for i in y_predicted]
  y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(x):
  return 1/(1+np.exp(-x))
```

```python
def gradient_descent(age,affordibility,y_true,epoch):
  w1=w2=1
  bias=0
  rate=0.001
  n=len(age)
  for i in range(epoch):
    weighted_sum=w1*age+w2*affordibility+bias
    y_pred=sigmoid_numpy(weighted_sum)
    loss=log_loss(y_true,y_pred)
    w1=w1-rate*np.mean(np.dot(np.transpose(age),(y_pred-y_true)))
    w2=w2-rate*np.mean(np.dot(np.transpose(affordibility),(y_pred-y_true)))
    bias=bias-rate*np.mean(y_pred-y_true)
    print(f'Epoch: {i},age:{w1},affordibilty :{w2},loss: {bias}')

gradient_descent(x_train_scale['age'],x_train_scale['affordibility'],y_train,50)

```
